## While loop

###  break, continue

````python
# break 
for i in range(10):
    if i==5: break # i5
    print(i)
print("EOP") # "EOP" 
````

````python
# continue    skip
for i in range(10):
    if i==5: continue # i5i
    print(i)
print("EOP") # "EOP" 
````

### else


````python
# 1
for i in range(10):
    print(i,)
else:
    print ("EOP")
````

````python
i=0
while i<10:
    print (i,)
    i += 1
else:
    print ("EOP") # breakelse block
````

````python
# -*- coding: utf-8 -*
import random                                 #    
guess_number = random.randint(1, 100)    # 1~100    
print ("  (1 ~ 100)")
users_input = int(input())                    #   
while (users_input is not guess_number):      #     
    if users_input > guess_number:            #    
        print ("  ")
    else:                                     #    
        print ("  ")
    users_input = int(input())                #    
else: print (". ", "  ", users_input , "") #  
````

````python
print ("   (1~9)?")
x = 1
while (x is not 0):                                           # (:0)
    x = int(input())
    if x == 0: break
    if not(1 <= x <= 9):
        print (" ", "1 9  ")
        continue
    else:
        print (" " + str(x) +" .")       
        for i in range(1,10):                                  # loop
            print (str(x) + " X " + str(i) + " = " + str(x*i))
        print ("   (1~9)?")
print ("  ")
````
